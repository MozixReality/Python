```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
plt.axis("equal")
fig = plt.gcf();
fig.set_size_inches(20, 20)
pi = np.linspace(0, 2*math.pi)

#face
r = 50;
face_x = r * np.cos(pi);
face_y = r * np.sin(pi);
plt.plot(face_x, face_y, color='black', linewidth=20)

#mouth --up
pi = np.linspace(1*math.pi, 1.9*math.pi);
r = 10;
mouth_xup = r * np.cos(pi)+10;
mouth_yup = r * np.sin(pi)-10;
plt.plot(mouth_xup, mouth_yup, color='black', linewidth=15)
plt.plot(-mouth_xup, mouth_yup, color='black', linewidth=15)

#mouth --down
t = np.linspace(-12, 12);
mouth_xdn = t
mouth_ydn = 1/10*t*t-35
plt.fill(mouth_xdn, mouth_ydn, color='#EA0000')
plt.plot(mouth_xdn, mouth_ydn, color='black', linewidth=15)

#inner mouth
pi = np.linspace(1*math.pi, 1.56*math.pi);
r = 10;
mouth_xup = r * np.cos(pi)+10;
mouth_yup = r * np.sin(pi)-10;
plt.fill_between(mouth_xup, mouth_yup, -21, color = '#EA0000')
plt.fill_between(-mouth_xup, mouth_yup, -21, color = '#EA0000')


#inner face
pi = np.linspace(0, 2*math.pi)
r = 40;
face_x = r * np.cos(pi);
face_y = r * np.sin(pi);
plt.fill(face_x, face_y, color='#FF1E00', alpha = 0.85)

#eyes
r = 8;
eye_x = r * np.cos(pi)+20;
eye_y = r * np.sin(pi)+10;
plt.fill(eye_x, eye_y, color='black');
plt.fill(-eye_x, eye_y, color='black');

```
