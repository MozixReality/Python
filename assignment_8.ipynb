```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

```

# 首先在二維平面找出一條 affine subspace

```python
x = np.array([12, 27])
vec = np.array([0.2, 0.3])
V = []

for i in range(-20, 20):
    point = x + i*vec
    V.append(point)
    plt.scatter(point[0], point[1], color = 'limegreen')
```

# 接著定義線性轉換矩陣 A 以及 一點 x0

```python
A = np.array([[0.2, 0.3],
              [0.7, -0.5]])
x0 = np.array([2, 25])
W = []

for i in range(40):
    x = A[0][0]*V[i][0] + A[0][1]*V[i][1]
    y = A[1][0]*V[i][0] + A[1][1]*V[i][1]
    point = np.array([x,y]) + x0
    W.append(point)
    plt.scatter(point[0], point[1], color = 'lightskyblue')
```

# 來 我畫給你看

```python
for i in range(40):
    plt.scatter(V[i][0], V[i][1], color = 'limegreen')
    plt.scatter(W[i][0], W[i][1], color = 'lightskyblue')
```
